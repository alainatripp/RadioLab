```python
import numpy as np
import matplotlib.pyplot as plt

loaded = np.load("100khz.npz")
data = loaded["data"]
fs   = loaded["fs"]

x = data[1].astype(float) ### Choosing not garbage data with 1 not 0

# Remove DC offset
x = x - np.mean(x)

N = len(x)

# FFT
X = np.fft.fft(x)
f = np.fft.fftfreq(N, d=1/fs)

# Power spectrum
P = np.abs(X)**2

# Plot only positive frequencies
mask = f > 0

plt.plot(f[mask]/1e3, P[mask])
plt.xlabel("Frequency (kHz)")
plt.ylabel("Power")
plt.title("Power Spectrum (100 kHz)")
plt.xlim(0, 1500)   # only up to Nyquist
plt.show() 
```

```python
# 1 Mhz with noise real

import numpy as np
#import seaborn as sns
from scipy.optimize import curve_fit
from scipy.stats import norm
import matplotlib.pyplot as plt

## fit histogram to a gaussian fit 
# use as the import package\
def gaussian_function(x, amplitude, mean, std_dev, offset):
    return amplitude * np.exp(-((x - mean) / std_dev)**2 / 2) + offset

# array from FFT power
data = X.real


# Generate histogram counts and bin edges
counts, bin_edges = np.histogram(data, bins=150)

# Calculate bin centers from bin edges
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

# Provide initial guesses for parameters (amplitude, mean, std_dev, offset)
# Good guesses can improve fit success; here we use an estimate from the data
initial_guesses = [max(counts), np.mean(data), np.std(data), 0]

# Perform the fit
# `popt` contains the optimized parameters
# `pcov` contains the estimated covariance of the parameters (used for error estimation)
popt, pcov = curve_fit(gaussian_function, bin_centers, counts, p0=initial_guesses, maxfev=10000)


# Extract the optimized parameters
fit_amplitude, fit_mean, fit_std_dev, fit_offset = popt

# Generate smooth x-values for the fitted curve
x_values = np.linspace(min(bin_centers), max(bin_centers), 1000)
# Calculate corresponding y-values using the optimized parameters
y_fit = gaussian_function(x_values, *popt)

# Plot the original histogram
plt.hist(data, bins=150, label='Original Histogram Data', alpha=0.6)

# Plot the fitted Gaussian curve
plt.plot(x_values, y_fit, 'r-', label=f'Gaussian Fit:\n mean={fit_mean:.2f}, std={fit_std_dev:.2f}')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Gaussian Fit to Histogram Data')
plt.xlim(-1950, 1950)
plt.legend()
plt.grid()
plt.show()


"""
# Labels and title
plt.title('Gaussian Noise compared with Ideal Gaussian')
#sns.histplot(data, kde=True, bins=150, color='purple', label='Ideal Gaussian')
plt.hist(X, bins=150, label='Gaussian Noise') ## changing bins to make it more gaussian looking
plt.xlim(-1950, 1950) ## and changing xlim for gaussian-aity
plt.legend()
plt.grid()
plt.show()
"""
```

```python
# 1600 khz w  noise

loaded = np.load("1600khzNoise.npz") 
data = loaded["data"]
fs   = loaded["fs"]

x = data[3].astype(float)
print(fs)

# Remove DC offset
x = x - np.mean(x)

N = len(x)

# FFT
X = np.fft.fft(x)
f = np.fft.fftfreq(N, d=1/fs)

# Power spectrum
P = np.abs(X)**2

# Plot only positive frequencies
mask = f > 0

plt.semilogy(f[mask]/1e3, P[mask])
plt.xlabel("Frequency (kHz)")
plt.ylabel("Power")
plt.title("Power Spectrum (1600 kHz)")
plt.xlim(0, 1500)   # only up to Nyquist
plt.show() 

plt.plot(X)
plt.show()
plt.plot(x)
plt.show()

#counts = len(X) # the stupid code is saying X isnt an array.
# plt.bar(counts, bins)
# plt.show()
#plt.hist(X, bins=1000, weights=counts)
plt.hist(X.real, bins=50) ## changing bins to make it more gaussian looking (add X.real?)
plt.xlim(-950, 950) ## and changing xlim for gaussian-aity
```
